```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets 
def get_iris_df():
  ds = sklearn.datasets.load_iris()
  df = pd.DataFrame(ds['data'],
    columns = ds['feature_names'])
  code_species_map = dict(zip(
    range(3), ds['target_names']))
  df['species'] = [code_species_map[c]
    for c in ds['target']]
  return df
df = get_iris_df()

# Load existing data from sklearn. Load data with panda lib into local dataset. Lastly call
# the local function to store data for the session.
# All of this code comes from the handbook for this module and was only tested as was said that
# any data may be used to explain te graphs. 
```

```python
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='pie', fontsize=20,figsize=(10,10))
plt.ylabel(var, horizontalalignment='left')
plt.title('Breakdown for ' + var, fontsize=25)
plt.show()
# Sum specific data by a species and plot the data on a pie chart
```

```python
df.plot(kind='hist', subplots=True, layout=(2,2), figsize=(10,10))
plt.suptitle('Iris Histograms', fontsize=20)
plt.show()

# plot data on 2 different species in terms of their length and width. 
# to understanding the x axis is the length or width and the y axis is the amount of it
```

```python
for spec in df['species'].unique():
  forspec = df[df['species']==spec]
  forspec['petal length (cm)'].plot(
    kind='hist', alpha=0.4, label=spec)
plt.legend(loc='upper right')
plt.suptitle('Petal Length by Species')
plt.show()
# Collect and determine dataset for data on all species and display them on a histogram 
# In terms of their length and amount in that length. 
    
```

```python
col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i: i% 50)
df.pivot('ind','species')[col].plot(kind='box')
plt.show()

# Get data in terms of mediums and high low values and area of most data and plot it on pivot table
```

```python
df.plot(kind="scatter",
    x="sepal length (cm)", y="sepal width (cm)")
plt.title("Length vs Width")
plt.show()

# Plot the scattered data in terms of lenth towards the sepal length of all species
```

```python
df.plot(kind="hexbin",
    x="sepal width (cm)", y="sepal length (cm)")
plt.show()

# not sure why the data is represented but I conclude that it shows the working of a temp plot
```

```python
import urllib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Get raw CSV data from the web
URL = ("http://ichart.finance.yahoo.com/" +   
    "table.csv?s=GOOG&c=2000")
dat = urllib.urlopen(URL).read()
open('foo.csv','w').write(dat)
# Make DataFrame, w timestamp as the index
df = pd.read_csv('foo.csv')
df.index = df['Date'].astype('datetime64')
df['LogClose'] = np.log(df['Close'])
df['Close'].plot()
plt.title("Normal Axis")
plt.show()
df['Close'].plot(logy=True)
plt.title("Logarithmic Axis")
plt.show()

# Was unable to collect data for this type of plot. But in this data is collected from
# an internet source and plotted in certain time configuration and plotted in log on y axis and normal
```
